```python
dfx = pd.read_excel('data/online_retail.xlsx')
dfx.head()
```

```python
dfx.columns
```

```python
dfx.Country.unique()
```

```python
dfx['Description'] = dfx['Description'].str.strip()
dfx['Description']
```

```python
dfx.dropna(axis = 0, subset =['InvoiceNo'], inplace = True)
dfx['InvoiceNo'] = dfx['InvoiceNo'].astype('str')
dfx['InvoiceNo'] 
```

```python
dfx = dfx[~dfx['InvoiceNo'].str.contains('C')]
dfx
```

```python
# Transactions done in France
basket_France = (dfx[dfx['Country'] =="France"]
          .groupby(['InvoiceNo', 'Description'])['Quantity']
          .sum().unstack().reset_index().fillna(0)
          .set_index('InvoiceNo'))
basket_France
```

```python
# Transactions done in the United Kingdom
basket_UK = (dfx[dfx['Country'] =="United Kingdom"]
          .groupby(['InvoiceNo', 'Description'])['Quantity']
          .sum().unstack().reset_index().fillna(0)
          .set_index('InvoiceNo'))
basket_UK
```

```python
# Transactions done in Portugal
basket_Por = (dfx[dfx['Country'] =="Portugal"]
          .groupby(['InvoiceNo', 'Description'])['Quantity']
          .sum().unstack().reset_index().fillna(0)
          .set_index('InvoiceNo'))
basket_Por
```

```python
basket_Sweden = (dfx[dfx['Country'] =="Sweden"]
          .groupby(['InvoiceNo', 'Description'])['Quantity']
          .sum().unstack().reset_index().fillna(0)
          .set_index('InvoiceNo'))
basket_Sweden
```

```python
def hot_encode(x):
    if(x<= 0):
        return 0
    if(x>= 1):
        return 1
```

```python
# Encoding the datasets
basket_encoded = basket_France.applymap(hot_encode)
basket_France = basket_encoded
basket_France
```

```python
basket_encoded = basket_UK.applymap(hot_encode)
basket_UK = basket_encoded
  
basket_encoded = basket_Por.applymap(hot_encode)
basket_Por = basket_encoded
  
basket_encoded = basket_Sweden.applymap(hot_encode)
basket_Sweden = basket_encoded
```

```python
from mlxtend.frequent_patterns import apriori, association_rules
```

```python
frq_items = apriori(basket_France, min_support = 0.05, use_colnames = True)
frq_items
```

```python
# Collecting the inferred rules in a dataframe
rules = association_rules(frq_items, metric ="lift", min_threshold = 1)
rules = rules.sort_values(['confidence', 'lift'], ascending =[False, False])
rules.head()
```

```python
frq_items = apriori(basket_UK, min_support = 0.01, use_colnames = True)
rules = association_rules(frq_items, metric ="lift", min_threshold = 1)
rules = rules.sort_values(['confidence', 'lift'], ascending =[False, False])
rules.head()
```
